```python
import pandas as pd
import numpy as np

def FillValue():
   
    df=pd.read_excel('Data_Train.xlsx')
    if df.empty:
        return df
    
    for column in df.columns:
        if df[column].dtype in [np.int64, np.float64]:
            mean = df[column].mean()
            df[column].fillna(mean, inplace=True)
        else:
            df[column].fillna(df[column].mode()[0], inplace=True)
        
    return df
```

```python
print("FILL all  MISSING VALUES with their mean median mode  ")
df_filled =  FillValue()
print(df_filled)
```
